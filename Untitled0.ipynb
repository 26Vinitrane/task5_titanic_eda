```python
import seaborn as sns
import matplotlib.pyplot as plt

numeric_df = df.select_dtypes(include=['float64', 'int64'])


plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of Numeric Features")
plt.show()


pairplot_df = df[['Age', 'Fare', 'Survived']].dropna()

sns.pairplot(pairplot_df, hue='Survived')
plt.show()

**Observation:**

HEATMAP CORRELATION
- `Fare` and `Pclass` have a negative correlation, meaning higher class passengers paid more.
- `Survived` has a slight positive correlation with `Fare` and negative with `Pclass`, indicating wealthier passengers had better chances of survival.


**Observation:**

PAIRPLOT
- Survivors tend to cluster around certain age and fare values.
- Theres a visible separation between survival classes, although its not very sharp  suggesting more features might be needed for classification.



```
